### Creating DataFrame

```python
# using lists
# with data columns like 
# IQ    Marks    Package
student_data = [
    [100, 80, 10],
    [90, 70, 7],
    [120, 100, 14],
    [80, 50, 2]
]

student_df = pd.DataFrame(student_data, columns=['iq', 'marks', 'package'])
student_df
```

```python
# using a dictionary

student_dict = {
    'name': ['Sajjad', 'Saad', 'Shahrukh', 'Salman', 'Hrithik', 'Priyanka'],
    'iq': [100, 90, 120, 80, 0, 0],
    'marks': [80, 70, 100, 50, 0, 0],
    'package': [10, 7, 14, 2, 0, 0]
}

students = pd.DataFrame(student_dict)
students.set_index('name', inplace=True)
students
```
