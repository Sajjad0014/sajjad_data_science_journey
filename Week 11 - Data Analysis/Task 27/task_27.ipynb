```python
count = metadata['totalCount'] // 5
count
```

```python
import time

num_iters = 1

for i in range(1, count):
    if num_iters == 20:
        break

    url = "https://wft-geo-db.p.rapidapi.com"
    page = url + next_url

    response = requests.request("GET", page, headers=headers)
    # print(response.json())

    links = response.json()['links']

    for item in links:
        if item['rel'] == 'next':
            next_url = item['href']

    data.extend(response.json()['data'])

    metadata = response.json()['metadata']

    num_iters += 1

    time.sleep(3)
```

7. actual price

```python
from selenium import webdriver
import time
import random

final = pd.DataFrame()

for j in range(1, 9):
    print(j)

    image_url = []
    name_of_image = []
    avg_ratings = []
    total_ratings = []
    total_reviews = []
    disc_prices = []
    actual_prices = []

    flipkart_url = f"https://www.flipkart.com/search?q=smartphones&page={j}"

    driver = webdriver.Chrome()
    driver.get(flipkart_url)
    page_content = driver.page_source
    driver.quit()

    soup = BeautifulSoup(page_content, 'lxml')

    formatted_html = soup.prettify()

    for phone in soup.findAll('img', class_='DByuf4')[:24]:
        image_url.append(phone.get('src'))
        name_of_image.append(phone.get('alt'))

    for phone in soup.find_all('div', class_='XQDdHH')[:24]:
        avg_ratings.append(phone.text)

    for phone in soup.find_all('span', class_="Wphh3N")[:24]:
        total_ratings.append(phone.text.split()[0])
        total_reviews.append(phone.text.split()[3])

    for phone in soup.find_all('div', class_="Nx9bqj _4b5DiR"):
        disc_prices.append(phone.text)

    for phone in soup.find_all('div', class_="yRaY8j ZYYwLA"):
        actual_prices.append(phone.text)

    d = {
    "image_urls": image_url,
    "name_of_image":name_of_image,
    "avg_ratings":avg_ratings,
    "total_ratings":total_ratings,
    'total_reviews':total_reviews,
    "disc_prices":disc_prices,
    "actual_prices":actual_prices
    }
    
    df = pd.DataFrame(d)
    
    final = final._append(df, ignore_index=True)
    time.sleep(random.uniform(1, 5))
```
