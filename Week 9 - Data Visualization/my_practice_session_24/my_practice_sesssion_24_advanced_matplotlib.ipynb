### Annotations

```python
batters = pd.read_csv('batter.csv')
print(batters.shape)
batters.head()
```

```python
sample_df = batters.head(100).sample(25, random_state=3)
print(sample_df.shape)
sample_df.head()
```

```python
plt.scatter(sample_df.avg, sample_df.strike_rate)
```

```python
x = [1, 2, 3, 4]
y = [5, 6, 7, 8]

plt.scatter(x, y)
plt.text(1, 5, 'Point 1')
plt.text(2, 6, 'Point 2')
plt.text(3, 7, 'Point 3')
plt.text(4, 8, 'Point 4', fontdict={'size': 12, 'color': 'yellow'})
```

```python
plt.figure(figsize=(18, 10))
plt.scatter(sample_df.avg, sample_df.strike_rate, s=sample_df.runs * 0.4)
plt.xlabel('Average')
plt.ylabel('Strike Rate')
for i in range(sample_df.shape[0]):
    plt.text(sample_df.avg.values[i], sample_df.strike_rate.values[i], sample_df.batter.values[i])
```

### Horizontal and Vertical Lines

```python
plt.figure(figsize=(18, 10))
plt.scatter(sample_df.avg, sample_df.strike_rate, s=sample_df.runs * 0.4)
plt.xlabel('Average')
plt.ylabel('Strike Rate')

plt.axhline(130, color='red')
plt.axhline(140, color='green')
plt.axvline(30, color='red')

for i in range(sample_df.shape[0]):
    plt.text(sample_df.avg.values[i], sample_df.strike_rate.values[i], sample_df.batter.values[i])
```

```python
fig = plt.figure()

# 2 rows 1st col and 1st graph
ax1 = fig.add_subplot(221)
ax1.scatter(batters.avg, batters.strike_rate, )

# 2 rows 2nd col and 2nd graph
ax2 = fig.add_subplot(2, 2, 2)
ax2.scatter(batters.avg, batters.runs)

# 2nd row 1st col and 1st graph
ax3 = fig.add_subplot(223)
ax3.hist(batters.strike_rate)

# 2nd row 2nd col and 2nd graph
ax4 = fig.add_subplot(2, 2, 4)
ax4.hist(batters.runs)
plt.show()
```

### 3D Surface Plot

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

x
```

```python
xx, yy = np.meshgrid(x, y)
print(xx.shape)
print(yy.shape)
```

```python
z = xx ** 2 + yy ** 2
z
```

```python
fig = plt.figure(figsize=(12, 8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx, yy, z, cmap='viridis')
fig.colorbar(p)
```

## Contour graph

```python
z = xx ** 2 + yy ** 2
z.shape
```

```python
fig = plt.figure(figsize=(7, 5))

ax = plt.subplot()

p = ax.contour(xx, yy, z, cmap='viridis')
fig.colorbar(p)
```

```python
## Contour f

fig = plt.figure(figsize=(7, 5))

ax = plt.subplot()

p = ax.contourf(xx, yy, z, cmap='viridis')
fig.colorbar(p)
```

```python
z = np.cos(xx) + np.sin(yy)
z
```

```python
fig = plt.figure(figsize=(12, 8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx, yy, z, cmap='viridis')
fig.colorbar(p)
```

```python
# z = np.cos(xx) + np.sin(yy)
fig = plt.figure(figsize=(7, 5))
ax = plt.subplot()
p = ax.contourf(xx, yy, z, cmap='viridis')
fig.colorbar(p)
```

```python
# pie -> single and multiple
df = pd.DataFrame(
    {
        'batsman': ['Dhawan', 'Rohit', 'Kohli', 'SKY', 'Pandya', 'Pant'],
        'match1': [120, 90, 35, 45, 12, 10],
        'match2': [0, 1, 123, 130, 34, 45],
        'match3': [50, 24, 145, 45, 10, 90]
    }
)

df.head()
```

```python
df['match1'].plot(kind='pie', labels=df.batsman.values, autopct='%0.1f%%')
```

```python
# multiple pie charts
df[['match1', 'match2', 'match3']].plot(kind='pie', subplots=True, figsize=(15, 8), labels=df.batsman.values,
                                        autopct='%0.1f%%')
```
