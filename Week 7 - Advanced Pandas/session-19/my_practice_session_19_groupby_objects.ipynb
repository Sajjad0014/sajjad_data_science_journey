```python
# Find the total number of groups --> len
len(movies.groupby('Genre'))
```

```python
movies['Genre'].nunique()
```

```python
# how to see the number of rows in each group
movies.groupby('Genre').size()
```

```python
movies['Genre'].value_counts()
```

```python
movies.groupby('Genre').last()
```

```python
# agg method on different columns
# passing dict
movies.groupby('Genre').agg(
    {
        'Runtime': 'mean',
        'IMDB_Rating': 'mean',
        'No_of_Votes': 'sum',
        'Gross': 'sum',
        'Metascore': 'min'
    }
)
```

```python
# agg method on different columns
# passing dict
movies.groupby('Genre').agg(
    {
        'Runtime': ['mean', 'min'],
        'IMDB_Rating': 'mean',
        'No_of_Votes': 'sum',
        'Gross': 'sum',
        'Metascore': 'min'
    }
)
```

```python
# passing list of aggregated functions
# applies on every numeric column
numeric_movies = movies.select_dtypes(include='number')
numeric_movies.groupby(movies['Genre']).agg(['min', 'max', 'mean', 'median'])
```

```python
# Adding both the syntax
numeric_movies.groupby(movies['Genre']).agg(
    {
        'Runtime': ['min', 'max'],
        'IMDB_Rating': 'mean',
        'No_of_Votes': ['sum', 'mean'],
        'Gross': 'sum',
        'Metascore': 'min'
    }
)
```

```python
# looping on groups
# group is a string of names
# data is data of that group

for group, data in movies.groupby('Genre'):
    print(group)
    print(type(group), type(data))
```

```python
# Find the highest rated movie of each genre
df = pd.DataFrame(columns=movies.columns)

# movies.groupby('Genre')['IMDB_Rating'].max()
for group, data in movies.groupby('Genre'):
    highest_rated = data[data['IMDB_Rating'] == data['IMDB_Rating'].max()]
    df = pd.concat([df, highest_rated], ignore_index=True)

print(df)
```

```python
# split (apply) combine
# apply -> builtin function

movies.groupby('Genre').apply(min)
```

```python
# find number of movies starting with A for each group

def foo(group):
    print(type(group))
    return group
```

```python
# here when we use apply() it keeps on sending the groups individually to this apply function
# You can see now how the type of group is a DataFrame

# .apply() works in 3 stages
# - It splits   (in our case groups when we use groupby())
# - applies custom logic
# - sends the results combined 

# split -> apply -> combine
movies.groupby('Genre').apply(foo)
```

```python
def foo(group):
    return group['Series_Title'].str.startswith('A').sum()
```

```python
# find number of movies starting with A for each group
movies.groupby('Genre').apply(foo)
```

```python
# find ranking of each movie in the group according to the IMDB score

def rank_movie(group):
    group['genre_rank'] = group['IMDB_Rating'].rank(ascending=False)
    return group
```

```python
movies.groupby('Genre').apply(rank_movie)
```

```python
# find normalized IMDB rating group wise

def normal(group):
    group['norm_rating'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min()) / (
            group['IMDB_Rating'].max() - group['IMDB_Rating'].min())

    return group
```

```python
movies.groupby('Genre').apply(normal)
```

```python
ipl = pd.read_csv('deliveries.csv')
ipl
```

```python
ipl.shape
```

```python
ipl.columns
```

```python
# find the top 10 batsman in terms of runs
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# find the batsman with max no of sixes
six = ipl[ipl['batsman_runs'] == 6]
six.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1).index[0]
```

```python
# find the batsman with most number of 4's and 6's in the last 5 overs
last_5_overs = ipl[ipl['over'] > 15]
last_5 = last_5_overs[(last_5_overs['batsman_runs'] == 4) | (last_5_overs['batsman_runs'] == 6)]

last_5.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1).index[0]
```

```python
# find V Kohli's record against all teams
kohli_df = ipl[ipl['batsman'] == 'V Kohli']
kohli_df.groupby('bowling_team')['batsman_runs'].sum().reset_index()
```

```python
# Create a function that can return highest score of any batsman
kohli_df = ipl[ipl['batsman'] == 'V Kohli']
kohli_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
def highest(batsman):
    temp_df = ipl[ipl['batsman'] == batsman]
    return temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]

highest('CH Gayle')
```

```python
X = pd.DataFrame([[1, 8.5, 85], [1, 9.12, 91.2]])
X
```

```python
A = np.dot(X.T, X)
A
```

```python
# Compute the inverse of A
A_inv = np.linalg.inv(A)

print("A (X^T X):")
print(A)
print("\nInverse of A:")
print(A_inv)
```
