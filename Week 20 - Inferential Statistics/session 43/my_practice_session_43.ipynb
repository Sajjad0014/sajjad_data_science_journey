```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 10  # number of trials
p = 0.5  # prob of success
size = 1000  # number of samples to generate. Performing 10 experiments (tossing the coin) 1000 times

binomial_dist = np.random.binomial(n, p, size)

# These values are 6 times Head came
# then 7 times head came
# like that for all 1000 times
print(binomial_dist[:30])
print(binomial_dist.shape)
```

```python
plt.hist(binomial_dist, density=True) 
plt.show()
```

## Central Limit Theorem

```python
num_samples = 1000
sample_size = 300
distribution_range = (0, 1)

# Generate samples from a uniform distribution
samples = np.random.uniform(distribution_range[0], distribution_range[1], (num_samples, sample_size))

samples
```

```python
# Calculate the sample means
sample_means = np.mean(samples, axis=1)

print(sample_means[:10])
print(sample_means.shape)
```

```python
plt.hist(sample_means, bins=30, density=True, edgecolor='black')
plt.title('Histogram of Sample Means')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.show()
```

```python
num_samples = 1000
sample_size = 30
distribution_range = (0, 1)

# Generate samples from a uniform distribution
samples = np.random.uniform(distribution_range[0], distribution_range[1], (num_samples, sample_size))

# Calculate the sample means
sample_means = np.mean(samples, axis=1)

plt.hist(sample_means, bins=30, density=True, edgecolor='black')
plt.title('Histogram of Sample Means')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.show()
```

```python
num_samples = 1000
sample_size = 50
lambda_param = 2

# Generate samples from a uniform distribution
samples = np.random.exponential(scale=1 / lambda_param, size=(num_samples, sample_size))

# Calculate the sample means
sample_means = np.mean(samples, axis=1)

plt.hist(sample_means, bins=30, density=True, edgecolor='black')
plt.title('Histogram of Sample Means')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.show()
```

```python
# set the parameters
num_samples = 1000
sample_size = 50

# Poisson Distribution parameters
poisson_lambda = 2

# Gamma Distribution parameters
gamma_shape = 2
gamma_scale = 1

# Binomial Distribution parameters
binomial_n = 10 
binomial_p = 0.5

# Generate samples from a uniform distribution
poisson_samples = np.random.poisson(lam=poisson_lambda, size=(num_samples, sample_size))
gamma_samples = np.random.gamma(shape=gamma_shape, scale=gamma_scale, size=(num_samples, sample_size))
binomial_samples = np.random.binomial(n=binomial_n, p=binomial_p, size=(num_samples, sample_size))

# Calculate the sample means
poisson_means = np.mean(poisson_samples, axis=1)
gamma_means = np.mean(gamma_samples, axis=1)
binomial_means = np.mean(binomial_samples, axis=1)

fig, axes = plt.subplots(3, 1, figsize=(8, 12))

axes[0].hist(poisson_means, bins=30, density=True, edgecolor='black')
axes[0].set_title('Histogram of Sample Means (Poisson Distribution)')
axes[0].set_xlabel('Sample Mean')
axes[0].set_ylabel('Density')

axes[1].hist(gamma_means, bins=30, density=True, edgecolor='black')
axes[1].set_title('Histogram of Sample Means (Gamma Distribution)')
axes[1].set_xlabel('Sample Mean')
axes[1].set_ylabel('Density')

axes[2].hist(binomial_means, bins=30, density=True, edgecolor='black')
axes[2].set_title('Histogram of Sample Means (binomial Distribution)')
axes[2].set_xlabel('Sample Mean')
axes[2].set_ylabel('Density')

plt.tight_layout()
plt.show()
```

```python
# set the parameters
num_samples = 10000
sample_size = 50

# Gamma Distribution parameters
gamma_shape = 2
gamma_scale = 1

# Calculate the theoretical mean and variance
theoretical_mean = gamma_shape * gamma_scale
theoretical_variance = (gamma_shape * gamma_scale ** 2)

samples = np.random.gamma(shape=gamma_shape, scale=gamma_scale, size=(num_samples, sample_size))

# Calculate the sample means 
sample_means = np.mean(samples, axis=1)

# Calculate the empirical mean and variance of the sample means 
empirical_mean = np.mean(sample_means)
empirical_variance = np.var(sample_means)

# Compare the theoretical and empirical values 
print(f"Theoretical mean: {theoretical_mean:.4f}")
print(f"Empirical mean: {empirical_mean:.4f}\n")
print(f"Theoretical variance/50: {theoretical_variance / 50:.4f}")
print(f"Empirical variance: {empirical_variance:.4f}")
```

## Central Limit Theorem Proof

```python
import pandas as pd
```

```python
deliveries = pd.read_csv('IPL_Ball_by_Ball_2008_2022.csv')
print(deliveries.shape)
deliveries.head()
```

```python
runs = deliveries.groupby('batter').sum()['batsman_run']
runs
```

```python
runs = runs.values
print(runs)
```

```python
import seaborn as sns

sns.displot(runs, kde=True)
```

```python
# n = 50/605, batsman taken for sample out of 605 batsman
# m = 1000, number of times sampling done
```

```python
np.random.choice(runs, size=10)
```

```python
m = 1000
n = 50
new_list = []

for i in range(m):
    x = np.random.choice(runs, size=n)
    new_list.append(x.mean())
```

```python
len(new_list)
```

```python
new_numpy = np.array(new_list)
```

```python
sns.displot(new_numpy, kde=True)
```

```python
print(
    f"The mean of the COMPLETE data is \t\t\t\t\t   {runs.mean():.4f}\nThe mean of 100 samples of means we have calculated is {new_numpy.mean()}\nYou Can notice how close they are to each other.\nThis proves Central Limit Theorem")
```

```python
print(
    f"We can do same with the Std Dev.\nThe Std Dev of the COMPLETE data is \t\t\t\t\t   {runs.std() / np.sqrt(50):.4f}\nThe Std. Dev of 100 samples of means we have calculated is {new_numpy.std():.4f}\nYou Can notice how close they are to each other.\nThis also proves Central Limit Theorem")
```

```python
print(
    f"We can do same with the Variance.\nThe variance of the COMPLETE data is \t\t\t\t\t   {runs.var() / 50:.4f}\nThe variance of 100 samples of means we have calculated is {new_numpy.var()}\nYou Can notice how close they are to each other.\nThis also proves Central Limit Theorem")
```
