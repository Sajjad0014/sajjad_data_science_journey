```python
import numpy as np
import pandas as pd
```

```python
flats = pd.read_csv('flats_cleaned_level_1.csv')
houses = pd.read_csv('house_cleaned_level_1.csv')

df = pd.concat([flats, houses], ignore_index=True)
df
```

```python
# Shuffling the data
df = df.sample(df.shape[0], ignore_index=True)
df.head()
```

```python
df.to_csv('gurgaon_properties.csv', index=False)
```
