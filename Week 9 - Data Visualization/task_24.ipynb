### **Problem-3:** Make a 3D *Surface* plot of this below mathematical equation.

$$z = |x| - |y|$$

````python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

xx, yy = np.meshgrid(x, y)
print(xx.shape)
z = np.abs(xx) - np.abs(yy)
z.shape
````

````python
fig = plt.figure(figsize=(12, 8))

ax = plt.subplot(projection='3d')
ax.plot_surface(xx, yy, z, cmap='winter')
ax.set_title('3D Surface Plot')
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.set_zlabel('z axis')

plt.show()
````

### `Problem-6` Use Pandas plot functions

* Line plot of closing value of top 5 Stocks in Year 2020.
* Take top 5 stocks based on total turnover in Year 2020

````python
# code here
df = pd.read_csv('nifty-50.csv')
print(df.shape)
df.head()
````

````python
df.Date = pd.to_datetime(df.Date)
df.Date.dtype
````

````python
df.set_index('Date', inplace=True)
````

````python
# df[df.Date.dt.year == 2020]
temp_df = df.query('Date.dt.year == 2020')
top_5_stocks = temp_df.groupby('Symbol')['Turnover'].sum().head().reset_index().Symbol.tolist()
top_5_stocks
````

````python
temp_df = temp_df.query('Symbol in @top_5_stocks')
temp_df.head()
````

````python
for name, group in temp_df.groupby('Symbol'):
    plt.plot(group.index, group.Close, label=name)

plt.legend()
plt.show()
````

````python
ax = temp_df.query('Symbol == "BAJAJ-AUTO"').plot(y='Close', use_index=True)

for company in top_5_stocks[1:]:
    temp_df.query(f'Symbol == {company}').plot(y='Close', labels=company, use_index=True, ax=ax)
````

### `Problem 9:` Create a surface plot and the 2 types of the contour plots of the below equation.

$$z = sin(\sqrt{x^2 + y^2})$$


````python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

xx, yy = np.meshgrid(x, y)
z = np.sin(np.sqrt(xx ** 2 + yy ** 2))
z.shape
````

````python
fig = plt.figure(figsize=(12, 8))

ax = plt.subplot(projection='3d')
ax.plot_surface(xx, yy, z, cmap='winter')
````

````python
ax = plt.subplot()
ax.contour(xx, yy, z, cmap='winter')
````
