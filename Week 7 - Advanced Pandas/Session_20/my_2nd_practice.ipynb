```python
# find orange cap holder of all the seasons
temp_df.groupby(['batsman', 'season'])['batsman_runs'].sum().reset_index().sort_values(['batsman_runs'], ascending=False).drop_duplicates(subset=['season'], keep='first').sort_values('season')
```
