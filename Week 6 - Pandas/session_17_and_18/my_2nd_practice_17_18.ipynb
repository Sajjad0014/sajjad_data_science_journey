```python
import pandas as pd

data = pd.Series([100, 200, 100, 300, 200, 300, 300, 300, 300, 400, 400])

print("Original Data:")
print(data)
```

```python
data = data.reset_index()
data.rename(columns={0:'data'},inplace=True)
data
```

```python
data['average_rank'] = data['data'].rank(method='average', ascending=False)
data.sort_values('average_rank')
```

```python
data['min_rank'] = data['data'].rank(method='min', ascending=False)
data.sort_values('min_rank')
```

```python
data['min_rank'] = data['min_rank'].apply(lambda x: int(x))
data.sort_values('min_rank')
```

```python
data['max_rank'] = data['data'].rank(method='max', ascending=False).apply(lambda x: int(x) if type(x) == float else 0)
data.sort_values('max_rank')
```

```python
data['dense_rank'] = data['data'].rank(method='dense', ascending=False).apply(lambda x: int(x))
data.sort_values('dense_rank', ascending=False)
```

```python
data['first_rank'] = data['data'].rank(method='first', ascending=False).apply(lambda x: int(x))
data.sort_values('first_rank', ascending=False)
```

```python
data.set_index('index', inplace=True)
data
```
