```python
import pandas as pd
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:,-1]
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor()
kfold = KFold(n_splits=5, shuffle=True, random_state=1)
scores = cross_val_score(knn, X, y, cv=kfold, scoring='r2')
```

```python
scores.mean()
# Not Good
```

## GridSearchCV

```python
from sklearn.model_selection import GridSearchCV

knn = KNeighborsRegressor()

param_grid = {
    'n_neighbors':[1, 3, 5, 7, 10, 12, 15, 17, 20],
    'weights':['uniform', 'distance'],
    'algorithm':['ball_tree', 'kd_tree', 'brute'],
    'p':[1, 2]
}

gcv = GridSearchCV(knn, param_grid, scoring='r2', refit=True, cv=kfold, verbose=2)
```

```python
gcv.fit(X, y)
```

```python
gcv.best_params_
```

```python
gcv.best_score_
```

```python
gcv.cv_results_
```

```python
pd.DataFrame(gcv.cv_results_).columns
```

```python
# Over here we need to check top results (plot) and analyze other results also to retrieve optimal parameters

pd.DataFrame(gcv.cv_results_)[['param_algorithm', 'param_n_neighbors', 'param_p', 'param_weights','mean_test_score']].sort_values('mean_test_score', ascending=False)
```

```python
# since we used refit=True, we can use gcv to predict new data
# gcv.predict(new_data)
```

## RandomizedSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV

rcv = RandomizedSearchCV(knn, param_grid, scoring='r2', refit=True, cv=kfold, verbose=2)

rcv.fit(X, y)
```
