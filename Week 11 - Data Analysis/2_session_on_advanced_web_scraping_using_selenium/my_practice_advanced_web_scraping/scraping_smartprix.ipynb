```python
with open('smartprix.html', 'r', encoding='utf-8') as f:
    html_code = f.read()
```

```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
soup = BeautifulSoup(html_code, 'lxml')
soup.prettify()
```

```python
containers = soup.find_all('div', class_="sm-product has-tag has-features has-actions")
len(containers)
```

```python
names = []
prices = []
ratings = []
sim = []
processor = []
ram = []
battery = []
display = []
camera = []
card = []
os = []

for i in containers:
    try:
        names.append(i.find('h2').text)
    except:
        names.append(np.nan)
    try:
        prices.append(i.find('span', {'class': 'price'}).text)
    except:
        prices.append(np.nan)

    score_div = i.find('div', class_="tags").find('div', class_='score rank-4-bg')

    # Check if score_div is found, then find <b> inside it
    if score_div:
        score = score_div.find('b')
        ratings.append(score.text if score else np.NaN)
    else:
        ratings.append(np.NaN)

    x = i.find('ul', {'class': 'sm-feat specs'}).find_all('li')
    try:
        sim.append(x[0].text)
    except:
        sim.append(np.nan)
    try:
        processor.append(x[1].text)
    except:
        processor.append(np.nan)
    try:
        ram.append(x[2].text)
    except:
        ram.append(np.nan)
    try:
        battery.append(x[3].text)
    except:
        battery.append(np.nan)
    try:
        display.append(x[4].text)
    except:
        display.append(np.nan)
    try:
        camera.append(x[5].text)
    except:
        camera.append(np.nan)
    try:
        card.append(x[6].text)
    except:
        card.append(np.nan)
    try:
        os.append(x[7].text)
    except:
        os.append(np.nan)
```

```python
print(len(name))
print(len(price))
print(len(ratings))
```

```python
df = pd.DataFrame({
    'model': name,
    'price': price,
    'rating': ratings,
    'sim':sim,
    'processor':processor,
    'ram':ram,
    'battery':battery,
    'display':display,
    'camera':camera,
    'card':card,
    'os':os
})
df.head()
```
