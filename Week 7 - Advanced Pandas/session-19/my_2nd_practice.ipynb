```python
def foo(group):
    return group['Series_Title'].str.startswith('A').sum()
```

```python
movies.groupby('Genre').apply(foo, include_groups=False)
```

```python
def rank_movie(group):
    group['genre_rank'] = group['IMDB_Rating'].rank(method='dense', ascending=False)
    return group
```

```python
movies.groupby('Genre').apply(rank_movie, include_groups=False)
```

```python
def normal(group):
    rating_col = group['IMDB_Rating']
    group['norm_rating'] = round((rating_col - rating_col.min()) / (rating_col.max() - rating_col.min()), 2)
    
    return group
```

```python
movies.groupby('Genre').apply(normal, include_groups=False)
```
