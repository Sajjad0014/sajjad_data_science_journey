#### Categorical distribution Plots
Univariate
##### Boxplot
A boxplot is a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile [Q1], median, third quartile [Q3] and “maximum”). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

![Box Plot](https://i.ytimg.com/vi/BE8CVGJuftI/maxresdefault.jpg)


```python
sns.boxplot(tips, x='total_bill')
```

```python
sns.boxplot(tips, x='sex', y='total_bill')
```

```python
sns.catplot(tips, kind='box', x='day', y='total_bill', hue='sex')
```

```python
# single boxplot -> numerical col
sns.boxplot(tips, y='total_bill')
```

## Categorical Estimate Plot -> for central tendency
### Barplot
cat vs num, aggregated

When there are multiple observations in each category, it also uses bootstrapping to compute a confidence interval around the estimate, which is plotted using error bars

```python
# average total bill
sns.barplot(tips, x='sex', y='total_bill')
```

```python
# some issue with errorbar
import numpy as np

sns.barplot(tips, x='sex', y='total_bill', hue='smoker', estimator=np.min)
```

```python
sns.barplot(tips, x='sex', y='total_bill', errorbar=None)
```

```python
sns.barplot(tips, x='sex', y='total_bill')
```

```python
# point plot
# bar plot - bars
# connecting the errors bar
sns.pointplot(tips, x='sex', y='total_bill', estimator='mean')
```

```python
sns.barplot(tips, x='sex', y='total_bill', hue='smoker')
```

```python
sns.pointplot(tips, x='sex', y='total_bill', hue='smoker', errorbar=None)
```

### countplot
A special case of the bar plot is when you want show the number of observations in each category rather than computing a statistic for a second variable. This is similar to a histogram over a categorical, rather than quantitative, variable  

- just like histogram is for numerical values

- this count plot is for categorical values

```python
sns.countplot(tips, x='sex')
```

```python
sns.countplot(tips, x='sex', hue='day')
```

```python
# faceting using catplot
sns.catplot(tips, kind='box', x='sex', y='total_bill', col='smoker', row='time')
```

## Regression Plots
- regplot 
- lmplot

In the simplest invocation, both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression

```python
sns.scatterplot(tips, x='total_bill', y='tip')
```

```python
# axes level
# hue parameter is not available
sns.regplot(tips, x='total_bill', y='tip')
```

```python
sns.lmplot(tips, x='total_bill', y='tip', hue='smoker')
```

```python
# resid plot
sns.residplot(tips, x='total_bill', y='tip')
```

## Multiplots

#### A sceond way to plot Facet plots -> FacetGrid

```python
# Figure level -> relplot, displot, catplot and lmplot
sns.catplot(tips, x='sex', y='total_bill', kind='violin', col='day', row='time')
```

```python
g = sns.FacetGrid(tips, col='day', row='time', hue='smoker')
g.map(sns.violinplot, 'sex', 'total_bill')
g.add_legend()
```

### Plotting Pairwise Relationship (PairGrid Vs Pairplot)


```python

iris.head()
```

```python
sns.pairplot(iris, hue='species')
```

```python
g = sns.PairGrid(iris, hue='species')
g.map(sns.scatterplot)
```

```python
# map_diag -> map_offdiag
g = sns.PairGrid(iris, hue='species')
g.map_diag(sns.boxplot)
g.map_offdiag(sns.kdeplot)
```

```python
# map_diag -> map_upper -> map_lower
g = sns.PairGrid(iris, hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
g = sns.PairGrid(iris, hue='species', vars=['sepal_width', 'petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

### JointGrid Vs Jointplot

joint plot

['scatter', 'hist', 'hex', 'kde', 'reg', 'resid']

```python
sns.jointplot(tips, x='total_bill', y='tip', kind='scatter')
```

```python
sns.jointplot(tips, x='total_bill', y='tip', kind='kde', hue='sex')
```

```python
sns.jointplot(tips, x='total_bill', y='tip', kind='reg')
```

```python
g = sns.JointGrid(tips, x='total_bill', y='tip')
g.plot(sns.scatterplot, sns.histplot)
```

```python
g = sns.JointGrid(tips, x='total_bill', y='tip', hue='smoker')
g.plot(sns.kdeplot, sns.violinplot)
```
