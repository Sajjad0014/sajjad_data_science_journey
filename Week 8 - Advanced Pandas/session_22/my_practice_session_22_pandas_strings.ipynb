```python
import pandas as pd
import numpy as np
```

```python
# What are vectorized string operations
# When we apply operations on vectors
# rule - use .str.

arr = np.array([1, 2, 3, 4])
arr
```

```python
arr * 4
# This is vector operation
```

```python
# Problem in vectorized operations in vanilla python
lst = ['cat', 'mat', 'rat', None]

[i.startswith('c') for i in lst]
```

Along with this, list is very slow

Python noted this and made a solution, which can also work with faster speed

```python
# How pandas solved this issue?

lst = pd.Series(['cat', 'mat', 'rat', None])
# .str is string accessor
lst.str.startswith('c')

# fast and optimized for larger dataset
```

```python
# import titanic
df = pd.read_csv('titanic.csv')
df['Name']
```

```python
# Common functions 
# lower/upper/capitalize/title
df['Name'].str.title()
```

```python
# len
df['Name'][df['Name'].str.len() == 82].values[0]
```

```python
# strip
str1 = '                     Don                      '
str1.strip()
```

```python
df['Name'].str.strip()
```

```python
# most frequently used
# split -> get
df['Name'].str.split(',')
```

```python
df['lastname'] = df['Name'].str.split(',').str.get(0)
df.head()
```

```python
df['Name'].str.split(',').str.get(1).str.split(' ')
```

```python
df['Name'].str.split(',').str.get(1).str.strip().str.split(' ', n=1, expand=True)
```

```python
df[['title', 'firstname']] = df['Name'].str.split(',').str.get(1).str.strip().str.split(' ', n=1, expand=True)
df.head()
```

```python
df['title'].value_counts()
```

```python
# replace
# miss, Ms and Mlle is same
df['title'] = df['title'].str.replace('Ms.', 'Miss.')
df['title'] = df['title'].str.replace('Mlle.', 'Miss.')
df['title'].value_counts()
```

```python
# filtering
# startswith/endswith
df['firstname'].str.startswith('A')
```

```python
df[df['firstname'].str.startswith('A')]
```

```python
df.info()
```

```python
# isdigit/isalpha
df[df['firstname'].str.isdigit()]
```

```python
# Advance level filtering
# applying regular expression (regex)
# contains
# search john -> in all cases
df[df['firstname'].str.contains('john', case=False)]
```

```python
# find lastname with start and end char vowel
df[df['lastname'].str.contains('^[aeiouAEIOU].+[aeiouAEIOU]$')]
```

```python
# not a vowel
df[df['lastname'].str.contains('^[^aeiouAEIOU].+[^aeiouAEIOU]$')]
```

```python
# slicing
df['Name'].str[::-1]
```
