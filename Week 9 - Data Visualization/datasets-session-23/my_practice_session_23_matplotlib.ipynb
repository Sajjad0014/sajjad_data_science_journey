```python
# from a pandas DataFrame
batsman = pd.read_csv('sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'], batsman['V Kohli'])
```

```python
plt.plot(batsman['index'], batsman['RG Sharma'])
```

```python
# plotting multiple plots
plt.plot(batsman['index'], batsman['RG Sharma'])
plt.plot(batsman['index'], batsman['V Kohli'])
```

```python
# label title
plt.plot(batsman['index'], batsman['RG Sharma'])
plt.plot(batsman['index'], batsman['V Kohli'])

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# color (hex) and line (width and style) and marker (size)
plt.plot(batsman['index'], batsman['RG Sharma'], color='#e31818')
plt.plot(batsman['index'], batsman['V Kohli'], color='#188de3')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# linestyle
plt.plot(batsman['index'], batsman['RG Sharma'], color='#e31818', linestyle='dashed')
plt.plot(batsman['index'], batsman['V Kohli'], color='#188de3', linestyle='dashdot')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# linewidth
plt.plot(batsman['index'], batsman['RG Sharma'], color='#e31818', linestyle='solid', linewidth=5)
plt.plot(batsman['index'], batsman['V Kohli'], color='#188de3', linestyle='dashdot', linewidth=3)

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# marker D - Diamond, +, >, <, o
# markersize
plt.plot(batsman['index'], batsman['RG Sharma'], color='#e31818', linestyle='solid', marker='+', markersize=10)
plt.plot(batsman['index'], batsman['V Kohli'], color='#188de3', linestyle='dashdot', marker='D')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# legend -> location
plt.plot(batsman['index'], batsman['RG Sharma'], color='#e31818', linestyle='solid', marker='+', markersize=10,
         label='Rohit Sharma')
plt.plot(batsman['index'], batsman['V Kohli'], color='#188de3', linestyle='dashdot', marker='D', label='Virat Kohli')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
# best default
plt.legend(loc='lower right')
```

```python
# limiting  axes

price = [50000, 53000, 57000, 60500, 50000, 54000, 4550000]
years = [2015, 2016, 2017, 2018, 2019, 2020, 2024]

plt.plot(years, price)
```

```python
# we can trim the above graph
plt.plot(years, price)
plt.ylim(0, 90000)
plt.xlim(2015, 2019)
```

```python
# grid
plt.plot(batsman['index'], batsman['RG Sharma'], color='#e31818', linestyle='solid', marker='+', markersize=10,
         label='Rohit Sharma')
plt.plot(batsman['index'], batsman['V Kohli'], color='#188de3', linestyle='dashdot', marker='D', label='Virat Kohli')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend()

plt.grid()
```

```python
# show

plt.plot(batsman['index'], batsman['RG Sharma'], color='#e31818', linestyle='solid', marker='+', markersize=10,
         label='Rohit Sharma')
plt.plot(batsman['index'], batsman['V Kohli'], color='#188de3', linestyle='dashdot', marker='D', label='Virat Kohli')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend()
plt.grid()
plt.show()
```

image.png

- Bivariate Analysis

- Use case - Finding correlation

```python
# plt.scatter simple function
x = np.linspace(-10, 10, 50)

x
```

```python
y = 10 * x + 3 + np.random.randint(0, 300, 50)
y
```

```python
plt.scatter(x, y)
```

```python
# plt.scatter on pandas data
# slower technique
df = pd.read_csv('batter.csv')
df = df.head(50)
df.head()
```

```python
plt.scatter(df['avg'], df['strike_rate'], color='red', marker='D')
plt.title('Avg and Strike Rate Analysis of Top 50 Batsman')
plt.xlabel('Average')
plt.ylabel('Strike Rate')
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# size
plt.scatter(tips['total_bill'], tips['tip'], color='#188de3', s=tips['size'] * 15)
plt.title('Total Bill Vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
```

```python
# scatterplot using plt.plot
# this is a faster technique
plt.plot(tips['total_bill'], tips['tip'], 'o')
```

```python
# plt.plot Vs plt.scatter
# plt.plot is faster
```

### Pie Chart

image.png

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
# simple data
data = [23, 45, 100, 20, 49]
marks = ['English', 'Science', 'Maths', 'Social Studies', 'Hindi']
plt.pie(data, labels=marks)
plt.show()
```

```python
# dataset
df = pd.read_csv('gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'], labels=df['batsman'], autopct='%0.1f%%')
plt.show()
```

```python
plt.pie(df['batsman_runs'], labels=df['batsman'], autopct='%0.1f%%',
        colors=['#1f77b4', '#2ca02c', '#ff7f0e', '#d62728', '#9467bd', '#17becf']
        )
plt.show()
```

```python
plt.pie(df['batsman_runs'], labels=df['batsman'], autopct='%0.1f%%', explode=[0.1, 0, 0, 0.1, 0, 0], shadow=True)
plt.show()
```
