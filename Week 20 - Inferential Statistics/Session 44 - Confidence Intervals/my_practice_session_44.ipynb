## Margin of Error Vs Population Standard Deviation (Z Procedure)

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
# Set Parameters
sample_sizes = 30
confidence_level = 95

# Calculate the critical value for the z-score
critical_value = stats.norm.ppf((1 + confidence_level / 100) / 2)

critical_value
```

```python
# Generate an array of population standard deviations
population_std_devs = np.arange(1, 101)

population_std_devs
```

```python
# Calculate the margin of errors for each population standard deviation
margin_of_errors = critical_value * (population_std_devs / np.sqrt(sample_sizes))
margin_of_errors
```

```python
# Plot the relationship
plt.plot(population_std_devs, margin_of_errors)
plt.title('Relationship Between Margin of Error and Population Standard Deviation (Z Procedure)')
plt.xlabel('Population Standard Deviation')
plt.ylabel('Margin of Errors')
plt.grid(True)
plt.show()
```

```python
## Relationship between Margin of Error and Sample Size (Z Procedure)

# Set parameters 
population_std_devs = 20
confidence_level = 95

# Calculate the critical value for Z-score
critical_value = stats.norm.ppf((1 + confidence_level / 100) / 2)
critical_value
```

```python
# Generate an array of sample sizes
sample_sizes = np.arange(2, 101)
sample_sizes
```

```python
# Calculate the margin of error for each sample size
margin_of_errors = critical_value * (population_std_devs / np.sqrt(sample_sizes))
margin_of_errors
```

```python
# Plot the relationship
plt.plot(sample_sizes, margin_of_errors)
plt.title('Relationship betweeen the Margin of Error and Sample Size (Z - Procedure)')
plt.xlabel('Sample Size')
plt.ylabel('Margin of Errors')
plt.grid(True)
plt.show()
```

## Relationship between Margin of Error and Critical Value (Z-Procedure)

```python
# Set parameters
sample_size = 30
population_std_dev = 20 

# Generate an array of confidence levels from 50% to 99%
confidence_level = np.arange(50, 100)
confidence_level
```

```python
# Calculate the critical values (Z-scores) for each confidence level
critical_values = stats.norm.ppf((1 + confidence_level / 100) / 2)
critical_values
```

```python
# Calculate the margin of error for each critical value
margin_of_errors = critical_values * (population_std_dev / np.sqrt(sample_size))
margin_of_errors
```

```python
# Plot the relationship
plt.plot(confidence_level, margin_of_errors)
plt.title('Relationship between the Margin of Error and Critical Value (Z-proedure)')
plt.xlabel('Critical Value')
plt.ylabel('Margin of Error')
plt.grid(True)
plt.show()
```
